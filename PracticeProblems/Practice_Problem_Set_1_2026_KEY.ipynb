For practice creating functions, build your own functions to solve for the volume of a pyramid, cylinder, and torus. The formulas to solve the volume of these shapes are shown below:


Pyramid: $V = \frac{lwh}{3}$   
Cylinder: $V = \pi r^2h$   
Torus: $V = (\pi r^2)(2\pi R)$ 

Run these examples of functions to solve for volume to check your answers. 

```python
def pyramid(l,w,h):
    return (l*w*h)/3
```

```python
pyramid(3,6,5)
```

```python
def cylinder(r,h):
    pi=3.14159
    volume=pi*r**2*h
    return volume
```

```python
cylinder(2,12)
```

```python
def torus(r,R):
    pi=3.14
    volume=(pi*r**2)*(2*pi*R)
    return volume
```

```python
torus(3,8)
```
