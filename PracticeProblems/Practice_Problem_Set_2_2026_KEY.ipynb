Create two arrays, x and y, with four elements each. Concatenate them along x-axis (by columns).

```python
x = np.array([4,12,13,15])
y = np.array([-5,30,2,199])
np.hstack([x,y])
```

Concatenate x and y along y-axis (by rows) and save as z.

```python
z = np.vstack([x,y])
z
```

Output the middle two elements of the second row of z.

```python
z[1,1:3]
```

Split the z array into two 2x2 arrays.

```python
splitH1,splitH2 = np.hsplit(z,[2])
splitH1
splitH2
```
