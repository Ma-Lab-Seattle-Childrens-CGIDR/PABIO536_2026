```python
# Start with importing the packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy
from scipy import stats
import sklearn
# We're also going to tell Jupyter to use inline plotting instead of notebook plotting
# It basically means you don't have to use plt.show() in every cell
%matplotlib inline
```

For these practice problems, we will be using different Python datasets. You can learn more about these datasets and plotting tutorials by searching the dataset names online. 

```python
# Import a premade Python dataset: Iris
#This data sets consists of 3 different types of irises' (Setosa, Versicolour, and Virginica) petal and sepal length, stored in a 150x4 numpy.ndarray. 
#The rows being the samples and the columns being: Sepal Length, Sepal Width, Petal Length and Petal Width.
from sklearn.datasets import load_iris
Data = load_iris() #Iris is in an array, to see what it looks like you can run "Data" in a new cell
Data.target #the integers representing the species of each observation
Data.target_names #the encoding scheme for species: 0 = setosa, 1 = versicolor, 2 = virginica
```

```python
iris = pd.DataFrame(np.column_stack((Data.data, Data.target)), columns = Data.feature_names+['target']) #Convert Iris array into a dataframe
iris['Species'] = iris.target.replace(dict(enumerate(Data.target_names))) #Creates new column to show iris species
iris = iris.drop(columns = ['target'])
iris.head()
```

### <font color = blue> Plotting Practice:

1. Create a scatter plot to compare the sepal width distribution between different species of irises.

```python
sns.catplot(x="Species", y="sepal width (cm)", data=iris)
plt.xlabel("Iris Species")
plt.ylabel("Sepal Width (cm)")
```

2. Create a swarm plot to look at the distribution of petal width for each species of iris.

```python
sns.swarmplot(x="Species",y="petal width (cm)", data = iris)
plt.xlabel("Iris Species")
plt.ylabel("Petal Width (cm)")
```

3. Create a box plot to compare sepal length distribution of all iris species. 

```python
sns.boxplot(x="Species", y="sepal length (cm)", data=iris,linewidth=2, width = 0.5)
plt.xlabel("Iris Species")
plt.ylabel("Sepal Length (cm)")
```

4. Create a plot where all of the petal length values are overlayed on a box plot of the distribution for each iris species.

```python
sns.catplot(x="Species", y="petal length (cm)", data=iris,color = 'k',alpha = 0.5)
sns.boxplot(x="Species", y="petal length (cm)", data=iris,linewidth=2, width = 0.5)
plt.xlabel("Iris Species")
plt.ylabel("Petal Length (cm)")
```

5. Create a line and then a scatter plot to look at the relationship between petal length and width for the versicolor iris.

```python
sns.lineplot(x = "petal width (cm)",y="petal length (cm)",data=iris.iloc[50:100,:])
sns.catplot(x = "petal width (cm)",y="petal length (cm)",data=iris.iloc[50:100,:], color = "k")
```
