1. Let's say we have a bouqet of two dozen roses. A dozen are white, four are pink, and the rest are yellow. Calculate the probability of finding two or fewer yellow roses if we randomly choose six of the two dozen roses.

```python
M = 24 # size of total population
n = 8 # number of successes in the total population
N = 6 # number of tries
xfew = 2 # number of successes among the tries (selecting a yellow rose)

# calculates probability of finding 2 or fewer yellow roses
pfew = stats.hypergeom.cdf(xfew, M, n, N)
pfew
```

2. Now calculate the probability of finding three or more yellow roses when we randomly choose six of the two dozen roses.

```python
xmore = 3 # number of successes among the tries
# calculates the probability of finding x or
pmore = 1 - stats.hypergeom.cdf(xmore,M,n,N)
pmore
```
