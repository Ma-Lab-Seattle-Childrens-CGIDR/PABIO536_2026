## Conditionals: if statements

If statements execute a particular set of commands if a logical condition is met.

````python
# Basic if statement
a = 3
b = 5
if b > a:
    print('the condition is true: b is greater than a')
````

### elif and else
If-elif-else statements enable the execution of different commands based on whether different conditions are true.

````python
x = 0
if x < 0:
    print('negative number')
elif x > 0:
    print('positive number')
else:
    print('zero')
````

### Nesting multiple if statements

If you need to test multiple conditional statements, you can nest if statements within each other using `elif`.

````python
# function with embeded if statements
def divisibleby(myNum):
    if myNum % 2 == 0:
        print('number is even')
        if myNum % 3 == 0:
          print('number is divisible by 3')
````

````python
divisibleby(2)
````

````python
divisibleby(3)
````

````python
divisibleby(6)
````

Contrast this against using an `and` conditional operator:

````python
def divisiblebyAND(myNum):
    if myNum % 2 == 0 and myNum % 3 == 0:
        print('number is even')
        print('number is divisible by 3')
````

````python
divisiblebyAND(2)
````

````python
divisiblebyAND(3)
````

````python
divisiblebyAND(6)
````

## Iteration: For-loops
For-loops enable iterating for a specific number of iterations.

The basic loop:

````python
for loopingVar in range(1,5,1):
    print(loopingVar)
````

You can adjust the starting and stopping index of `range` by adjusting the input arguments. (More information about `range` can be found [here](https://www.w3schools.com/python/ref_func_range.asp)).

````python
for loopingVar in range(2,10,2):
    print(loopingVar)
````

Note, that you can also use your `loopingVar` to iterate over the elements of a list or a `numpy` ndarray.

### Nesting multiple for-loops:

For-loops can also be nested. The innermost for-loop gets iterated first.

````python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

# a nested for loop
for x in adj:
    for y in fruits:
        print(x, y)
````

### Nesting for-loops with conditional statements

More complex computations can be made by combining for-loops and if-statements.

Here's an example of how you can identify significant genes out of a list.

Note: we're also storing the significant values in new variables so that we can access them after the for-loop is complete.

````python
pvals = [0.01,0.03,0.07,0.08,0.6,0.02,0.04,0.01] #a list of p-values generated from some other calculation.
sigIndex = [] # Creating an empty list to store the index values of which elements are significant
sigPvals = [] # Creating an empty list to store the p-values that are significant
for p in range(0,len(pvals),1):
  if pvals[p] < 0.05:
    print('significant pvalue:',pvals[p],', iteration number', p+1)
    sigPvals.append(pvals[p])
    sigIndex.append(p)
````

````python
print(sigPvals)
print(sigIndex)
````

### <font color=blue> Optional Bonus: Using Continue and Break
If we combine for- and while- loops with if-statements, we can perform more complex computations. Combining iteration with conditionals also enables the use of these additional operations:

- **continue** - skip an iteration
- **break** - end the loop

#### Continue

````python
# for loop with continue to print out only odd numbers
for myVar in range(1, 11):
    if myVar % 2 == 0 :
        #pass
        continue

    print('This number = ' + str(myVar))
````

#### Break

````python
# For loop with a break
for myVar in range(1, 11):
    if myVar > 5 :
        #pass
        break

    print('This number = ' + str(myVar))
````

### <font color=blue> Optional Bonus: For-else loops

If you add an `else` to for-loops, it will execute at the end, after the for-loop has finished iterating.

````python
for loopingVar in range(0,4,1):
    print(loopingVar)
else:
    print('done with the iterations')
````

#### The effect of Break on For-else loops

If the `break` statement is activated, then statements in the `else` blocks of for-else loops will not be executed.

````python
# for-else loop with break
def forElse(myNum):
    for myVar in range(1,6):
        if myVar == myNum:
            print( 'if statement is satisfied, breaking out of the loop')
            break
        else:
            print('This number = ' + str(myVar))
            myVar += 1
    else:
        print( 'break statement is not executed, printing else block')
````

````python
forElse(3)
````

````python
forElse(10)
````

#### <font color=brown>Hands on practice</font>

Create a function that uses a for-else loop to assess whether an input number is prime. Have this function print out the number's first two factors and then break if the input number is not prime. Otherwise, have the function print out that the input number is prime.

*Hint:* The `%` operator enables the calculation of remainders.


````python
def primeAssessment(myNum):
  for i in range(2,myNum): #to iterate on the factors of the number
    if myNum % i == 0:      #to determine the first factor
      j = myNum / i #to calculate the second factor
      print(myNum,'equals ', i ,'* ', j)
      break #to move to the next number, the #first FOR
  else:        # else part of the loop
    print(myNum, 'is a prime number')
````

````python
primeAssessment(15)
````

### <font color=blue> Optional Bonus: While-loops
Sometimes you might encounter a situation in which you don't know how many times you need to iterate, but you do need to iterate a command until a condition is met. While-loops are suitable for executing this kind of function.

The basic while-loop:

````python
targetNum = 5
countVar = 1
while countVar <= targetNum:
    print(countVar)
    #countVar = countVar + 1
    countVar += 1
````
