## Libraries

Libraries are collections of functions and data types with useful properties that have been written by other people (so that you don't have to). Once you have these libraries installed on your computer, you will need to "`import`" the library into each Python session to use the functionality associated with these libraries.

To check the answer from our function, we're going to use a Python library called `statistics`, which includes a pre-written function for calculating sample standard deviations.

```python
import statistics
```

The syntax for calling a function from a library is `<name of library>.<name of function>`

```python
statistics.stdev([1,2,3])
```

The names of some libraries are long, so you can also give the libraries an alias:

```python
import statistics as stat
```

```python
stat.stdev([1,2,3])
```
