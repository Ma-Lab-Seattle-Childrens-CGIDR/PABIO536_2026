## Data Frame Slicing

````python
# creating some lists of information
fruits = ['apple','banana','cherry','mango']
color = ['red','yellow','red','yellow']
size = [10,20,3,12]
tropical = [False,True,False,True]

# creating a data frame from the lists
fruitsDF = pd.DataFrame({'Color': color, 'Size':size, 'Tropical Origin':tropical},index=fruits)
display(fruitsDF)
````

To slice with either `.loc[]` or `.iloc[]`, the convention is to use square brackets, and refer to the *rows* that you want, followed by a `,`, followed by the *columns* that you want: 

````python
fruitsDF.loc['apple','Color']
````

````python
fruitsDF.iloc[0,0:2]
````

In addition to explicitly listing the row/columns by name/index number, you can also slice the rows/columns that you want with a boolean reference.

````python
fruitsDF.loc[[True,False,True,False],'Color']
````

````python
fruitsDF.iloc[[True,False,True,False],0]
````

You can derive this boolean set by constructing a boolean query, and using the query to slice:

````python
print(fruitsDF['Color'] == 'red')
````

````python
fruitsDF.loc[fruitsDF['Color'] == 'red','Color']
````

The column and row slices are independent, so the boolean query does not have to come from the same column (or even data frame) as the column that you want to slice out.

````python
fruitsDF.loc[fruitsDF['Color'] == 'red',['Color','Tropical Origin']]
````

You can use Boolean and (`&`) / or (`|`) statements to add refinement to what you want to slice out. 

````python
fruitsDF.loc[(fruitsDF['Color'] == 'red') & (fruitsDF['Size'] > 5),['Color','Tropical Origin']]
````

## For loops and if statements

````python
adj = ["red", "big", "tasty"]
fruits = ["apple", "pomeogranate", "cherry"]
````

````python
# a nested for loop
for x in adj:
    print('here are the ', x, 'fruits:')
    for y in fruits:
        print(x, y)


````

````python
# a for loop with a conditional statement
for x in adj:
    print('here are the ', x, 'fruits:')
    for y in fruits:
        print(x, y)
        if x == 'tasty':
            print('yum')
````

````python
# a for loop with a conditional statement
for x in adj:
    print('here are the ', x, 'fruits:')
    if x == 'tasty':
        print('yum')
    for y in fruits:
        print(x, y)
````
