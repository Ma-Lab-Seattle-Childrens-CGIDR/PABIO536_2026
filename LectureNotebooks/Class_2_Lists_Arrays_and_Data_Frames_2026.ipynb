#### Removing elements from lists

You can remove specific elements that you don't want from lists, using either the `.remove()` method, or the `del()` function.

If you know which position, you want to remove, use `del`. The `del` function removes the element from a specific location (based on the slice that is input into the argument).

If you know what contents you want to remove (but not where it's located), use the `.remove()`. The `.remove()` will remove the first element in the list (from left to right) that matches the contents of the input argument.

```python
print(mixedlist) #reminder of what mixedlist currently contains
```

```python
# remove the second element
del(mixedlist[1])
print(mixedlist)
```

```python
# remove the first element that matches the contents of the input argument
mixedlist.remove(2)
print(mixedlist)
```

You can check the size of the list by using the function `len()`.

```python
len(mixedlist)
```

### Intersections and unions with `sets`
You might find it useful to tell what is in common to two lists (intersection), or what are the unique elements that are represented in both lists (union). One way of determining this us by using `sets`. Sets store unique elements, and can be used with the boolean/logical operator `&` (AND) to perform a set intersection and output the shared values between both lists. Using the `|` (OR) operator would give the union.

```python
print(mixedlist) #reminder of what mixedlist currently contains
```

```python
secondList = [2,3,'ten',True]
print(secondList)
```

```python
commonlist = list(set(mixedlist) & set(secondList)) #The "&" is the boolean/logical AND operator
print(commonlist)
```

```python
unionlist = list(set(mixedlist) | set(secondList)) #The "|" is the boolean/logical OR operator
print(unionlist)
```

Notice that the ordering of the elements in `unionlist` is not the same as the input lists.

## Using `numpy` and `pandas` to hold and manipulate data

While you can use lists to do some very basic data manipulation tasks, there are other data structures that in Python, including `numpy arrays` and `pandas data frames`, which both give a lot more flexibility and power to perform mathematical operations on the data contained in these objects.

`Numpy` introduces a new object for holding groups of variables: n-dimensional arrays of data (aka `ndarrays`).

After we introduce you to ndarrays we will switch to `pandas`, which is a wrapper for ndarrays that makes them much easier to use. Then we'll come back to `numpy` and show you some of the functions from `numpy` and `scipy`, which adds more complex mathematical and statistical functions to python.

### NumPy arrays: a new thing to hold other things

NumPy arrays (`ndarrays`) are essentially lists with two important restrictions:

 - An array can only hold **one** type of data
 - Arrays are somewhat unmutable:
     - you **can** change the contents of an ndarray
     - but you **can't** change the size or the data type of the array

Why would we want a list with extra restrictions? Speed, both in terms of time to compute and time to program.

The computer only needs to check the type of data once for the array, not once for each variable. This adds up when you have huge arrays like the results of an 'omics' experiment.

Being forced to organize your data makes it simpler to search, manipulate, and subset your data.

### Making a new `ndarray`
There are a number of ways to make `ndarrays`. You can import data from text files, convert a list to an `ndarray`, or with one of the `numpy` functions that builds basic `ndarray` types useful for data analysis.

```python
# The `numpy` function arange(start, stop, step) gives you an array of values
# between the start and stop (not including the stop) incremented by step
# The default step is 1
array1 = np.array([1,2,3,4])

print(array1)
```

A key difference between ndarrays and lists is that mathematical operators work with differently. We'll demonstrate this by converting an ndarray to a list using `ndarray.tolist()` and performing the same operations on both.

```python
#Convert a to a_list and then multiply both by two
a_list = array1.tolist()
print(array1)
print(a_list)
print(array1 * 2)
print(a_list * 2)
```

Another key difference is that ndarrays can be made into any number of dimensions.

```python
# Now let's make a 2D array (in this case, a 4 x 2 array)
array2 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(array2)
```

ndarray variables contain several "attributes" which describe the shape of the array:

```python
print(array2.size)  # number of elements
print(array2.ndim)  # number of dimensions
print(array2.shape) # number of elements per dimension (starts with rows, then columns)
```

### Subsetting

To extract subsets from arrays, you use the same slicing rules as for lists, and separate each dimension by using a "," sign:

```python
print(array2[:,1:3])
```

### Computation with `ndarrays`
`ndarrays` are particularly useful for performing math. Here's an example with summing.

```python
print(array1) #reminder of contents of array1
```

```python
array1multiply = np.multiply(array1,5) #multiply each element by 5
print(array1multiply)
```

```python
arrayNeg = -1 * array1
print(arrayNeg)
```

```python
absValue = np.abs(arrayNeg) #generates absolute value
print(absValue)
```

```python
array1log2 = np.log2(array1) # take log transform
print(array1log2)
```

You can also aggregate or accumulate mathematical calculations

```python
array1sum = np.sum(array1) #summing the elements of array1 together
print(array1sum)
```

```python
array1cumsum = np.cumsum(array1) #cumulative sum
print(array1cumsum)
```

### Import data into a DataFrame
Let's import some data to work with.

`Pandas` provides simple tools for importing from Excel, csv, or any other common data format:
 - `read_csv`
 - `read_excel`

We'll start by importing a table of metadata using `pd.read_csv()`.

This dataset comes from an [RNAseq experiment comparing melanoma cell lines from 2017](https://www.nature.com/articles/s41467-017-02353-y) published in Nature Communications.

```python
# Import comma-seperated data from a text file
df = pd.read_csv('~/LECTURE_MATERIALS/DataFiles/melanoma_CountsRaw.csv', index_col=0)
```

```python
# Take a look at the contents of the variable df using df.head()
df.head()
```

Pandas by default prefers to spreadsheet data organized such that individual samples/replicates are in represented in the rows, and different parameters/variables/properties are represented in the columns.

To take examine the data frame produced, `.shape` and `.head()` are useful:

```python
# Data frames store information about their shape (i.e. numbers of rows and columns) in the .shape attribute.
df.shape
```

### Subsetting

One of the powerful features of data frames is that rows and columns can be referred to by names or numbers:

```python
# Extracting data in all rows from the column named "Stage"
df['Stage']
```

```python
# Extracting data in all rows from the column named "Stage"
df.Stage
```

```python
# Extracting data in all columns from row "FM_1" (note that you need to add ".loc" in order to refer to the row names)
df.loc['FM_1']
```

To select multiple rows or columns, input a list:

```python
df.loc[['FM_1','FM_2','FM_3'],'Stage']
```

You can also subset data by number by using "`.iloc`" (zero-indexing also applies here)

```python
df.iloc[0:3,2]
```

To get Python to output the entire set of row and column names, you can use the `.index` and `.columns` attributes, respectively.

```python
# listing the row names
df.index
```

```python
# listing the column names
df.columns
```

Data frames can also be sliced by rows without explicitly naming the row names or row numbers. Instead, we can slice by testing whether a condition is fulfilled or not:

```python
# Finding out which rows match the stage: "primary melanocytes"
primary_indices = df.Stage == 'primary melanocytes'
print(primary_indices)
```

```python
# subsetting the rows based on this conditional set
primary_melanocytes_data = df[primary_indices]
primary_melanocytes_data
```

```python
# subsetting by conditionals in both rows and columns
primary_melanocytes_data2 = df.loc[df.Stage == 'primary melanocytes','cell_line']
primary_melanocytes_data2
```

You can driectly perform mathematical operations on numerical subsets of data frames:

```python
# adding the expression values in FM_2 and FM_3 for genes A1BG to A1CF
primary_melanocytes_data['A1BG'].iloc[1:3] + primary_melanocytes_data['A1CF'].iloc[1:3]
```

You can slice subsets of a data frame and convert back to a `numpy` ndarray or list by using:

*   `.to_numpy()` for ndarrays
*   `.tolist()` for lists.



```python
# using .to_numpy to convert the data in the column corresponding to the gene 'A1BG' in the primary_melanocytes_data data frame to a numpy array.

A1BGexpPrimary = primary_melanocytes_data['A1BG'].to_numpy()
print(A1BGexpPrimary)
```

You can also export dataframes back to Excel using the `.to_excel()` method.

```python
# exporting primary_melanocytes_data back to excel
primary_melanocytes_data.to_excel('primary_melanocytes_data.xlsx',index = False)
```

#### <font color=blue> Optional Bonus: Combining and Splitting arrays

For 1D and 2D arrays, the clearest way to concatenate arrays together is by using the `.hstack()` and `.vstack()` functions:

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])

# concatenate along x-axis (by columns)
np.hstack([x, y])
```

```python
# concatenate along the second (y-) axis (by rows) (axes are also zero-indexed).
np.vstack([x, y])
```

For 1D and 2D arrays, splitting the arrays into smaller pieces can be most clearly done using the `np.vsplit()` and `np.hsplit()` functions:

```python
print(array2)
```

```python
print(splitV1)
```

```python
print(splitV2)
```
